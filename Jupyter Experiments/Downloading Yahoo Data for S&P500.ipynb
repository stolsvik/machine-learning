```python
# Author: Endre St√∏lsvik, 2017-04-23
import json
import re
        
from urllib.parse import urlencode
from urllib.request import urlopen
API_URL = "https://en.wikipedia.org/w/api.php"

def get_wiki_text(title):
    data = {"action": "query", "prop": "revisions", "rvlimit": 1,
            "rvprop": "content", "format": "json", "titles": title}
    raw = urlopen(API_URL, urlencode(data).encode()).read().decode()
    res = json.loads(raw)
    return next(iter(res["query"]["pages"].items()))[1]["revisions"][0]["*"]

text = get_wiki_text("List of S&P 500 companies")
```

```python
lines = text.split('\n')

lines_i = iter(lines)

# Skip to the paragraph with the tickers
while (True):
    line = next(lines_i)
    if 'Component Stock' in line:
        break

# Iterate over the constituent stocks table
tickers = []
while (True):
    line = next(lines_i)
    # If we come to next paragraph (changes in index), we're done
    if ('==') in line:
        break
    # TO VIEW: print(line)
    match = re.match(r'.*\{\{\w*\|(\w*)', line)
    # Only parse lines with "{{XxxSymbol|TICKER}}", as those are the table rows.
    if (match == None):
        continue
    ticker = match.group(1)
    print("Ticker:", ticker)
    tickers.append(ticker)

print("Tickers length:", len(tickers))
```
